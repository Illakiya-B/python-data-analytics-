# Data Collection

## Objective

The purpose of this notebook is to collect historical daily market data for Bitcoin (BTC), Ethereum (ETH), Gold (GLD), and the S&P 500 Index using the Yahoo Finance API.

The collected datasets will serve as the primary data source for subsequent preprocessing, exploratory data analysis, statistical modeling, and financial analysis.

This notebook also performs an initial assessment of data completeness before storing the datasets in the project's raw data directory.

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

import yfinance as yf

from pathlib import Path
```

```python
# Create project directories if they do not already exist

RAW_DATA_PATH = Path("crypto_project/data/raw")
RAW_DATA_PATH.mkdir(parents=True, exist_ok=True)

print("Raw data directory is ready.")
```

```python
assets = {
    "Bitcoin": "BTC-USD",
    "Ethereum": "ETH-USD",
    "Gold": "GLD",
    "SP500": "^GSPC"
}
```

```python
#data downloading

market_data = {}

for asset_name, ticker in assets.items():

    print(f"Downloading {asset_name}...")

    df = yf.download(
        ticker,
        start="2019-01-01",
        end="2025-01-01",
        auto_adjust=True,
        progress=False
    )

    df.reset_index(inplace=True)

    market_data[asset_name] = df

print("Download Complete.")
```

```python
# Data Inspection

for asset_name, df in market_data.items():

    print("="*60)
    print(asset_name)

    display(df.head())

    print(df.info())
```

```python
# Data Quality Assessment

summary = []

for asset_name, df in market_data.items():

    summary.append({

        "Asset": asset_name,

        "Rows": len(df),

        "Columns": df.shape[1],

        "Missing Values": df.isna().sum().sum(),

        "Start Date": df["Date"].min(),

        "End Date": df["Date"].max()

    })

summary_df = pd.DataFrame(summary)

summary_df
```

```python
# Saving raw data

for asset_name, df in market_data.items():
     # Flatten MultiIndex columns if present
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    filename = asset_name.lower() + ".csv"
    df.to_csv(RAW_DATA_PATH / filename, index=False)

print("All raw datasets saved successfully.")
```

# Summary

The historical market data for Bitcoin, Ethereum, Gold, and the S&P 500 were successfully collected using the Yahoo Finance API.

The datasets were inspected for completeness and exported to the project's raw data directory. These files will be used in the next notebook for data cleaning and preprocessing.
