# Data Cleaning & Preprocessing

## Objective

The objective of this notebook is to clean, standardize, and integrate the historical datasets collected from Yahoo Finance into a single analysis-ready dataset.

The preprocessing steps ensure data consistency and improve the quality of subsequent exploratory analysis and statistical modeling.

```python
import pandas as pd
import numpy as np

from pathlib import Path
```

```python
RAW_DATA_PATH = Path("crypto_project/data/raw")

PROCESSED_DATA_PATH = Path("crypto_project/data/processed")

PROCESSED_DATA_PATH.mkdir(parents=True, exist_ok=True)
```

```python
bitcoin = pd.read_csv(RAW_DATA_PATH / "bitcoin.csv")

ethereum = pd.read_csv(RAW_DATA_PATH / "ethereum.csv")

gold = pd.read_csv(RAW_DATA_PATH / "gold.csv")

sp500 = pd.read_csv(RAW_DATA_PATH / "sp500.csv")
```

```python
datasets = {
    "Bitcoin": bitcoin,
    "Ethereum": ethereum,
    "Gold": gold,
    "SP500": sp500
}
```

```python
#dataset overview

for name, df in datasets.items():

    print("="*60)
    print(name)
    print(df.shape)
    display(df.head())
```

```python
#data type 
for name, df in datasets.items():

    print(f"\n{name}")

    print(df.dtypes)
```

```python
#missing value
missing_summary = pd.DataFrame()

for name, df in datasets.items():

    missing_summary[name] = df.isnull().sum()

missing_summary
```

```python
#duplicate record
for name, df in datasets.items():

    print(name)

    print("Duplicate Rows:", df.duplicated().sum())
```

```python
#date range for name, df in datasets.items():
for name, df in datasets.items():
    print(name)

    print(
        df["Date"].min(),
        "to",
        df["Date"].max()
    )
```

## DATA CLEANING

```python
# convert date
for df in datasets.values():

    df["Date"] = pd.to_datetime(df["Date"])
#sort data
for df in datasets.values():

    df.sort_values("Date", inplace=True)

    df.reset_index(drop=True, inplace=True)
```

```python
#data type 
for name, df in datasets.items():

    print(f"\n{name}")

    print(df.dtypes)
```

```python
#add asset column 
bitcoin["Asset"] = "Bitcoin"

ethereum["Asset"] = "Ethereum"

gold["Asset"] = "Gold"

sp500["Asset"] = "SP500"
```

## DATA INTEGRATION

```python
master_df = pd.concat(
    [bitcoin,
     ethereum,
     gold,
     sp500],
    ignore_index=True
)
master_df.sort_values(
    ["Date", "Asset"],
    inplace=True
)

master_df.reset_index(
    drop=True,
    inplace=True
)
```

```python
#Saving Master
master_df.to_csv(
    PROCESSED_DATA_PATH /
    "market_data_master.csv",
    index=False
)

print("Processed dataset saved successfully.")
```

# Summary

The raw datasets were successfully validated, cleaned, and integrated into a single master dataset.

The processed dataset is now standardized and ready for exploratory data analysis.
