# Research Question

**Has Cryptocurrency Decoupled from Traditional Financial Markets?**

This study examined the relationship between Bitcoin, Ethereum, Gold, and the S&P 500 between 2019 and 2024 using historical market data and financial analytics techniques.

The analysis focused on three key dimensions:

- Market Correlation
- Risk Characteristics
- Investment Performance

# Summary of the Analysis

The project followed a structured analytics workflow:

1. Collected historical daily market data using Yahoo Finance.
2. Cleaned and integrated datasets into a unified analysis-ready format.
3. Performed exploratory data analysis to understand market behaviour.
4. Engineered financial features such as daily returns, cumulative returns, rolling statistics, and log returns.
5. Conducted correlation analysis to evaluate relationships between cryptocurrencies and traditional assets.
6. Compared risk and performance using volatility, maximum drawdown, CAGR, and Sharpe Ratio.
7. Evaluated the proposed research hypotheses based on empirical evidence.

# Hypothesis Evaluation

| Hypothesis                      | Status              | Supporting Evidence                                         |
| ------------------------------- | ------------------- | ----------------------------------------------------------- |

| **H2 – Risk Characteristics**   | Accepted | Volatility, drawdown, recovery time                         |

# Key Findings

- Bitcoin generated the highest cumulative return over the study period.
- Ethereum exhibited the highest price volatility among all assets.
- Gold remained the most stable asset and acted as a relatively defensive investment.
- The S&P 500 demonstrated steady long-term growth with lower volatility than cryptocurrencies.
- Rolling correlation analysis indicated that relationships between cryptocurrencies and traditional assets changed over time rather than remaining constant.
- Market stress events, particularly the COVID-19 pandemic and the 2022 crypto downturn, significantly influenced asset behaviour.

# Business Insights

### Portfolio Diversification

Although cryptocurrencies may offer substantial returns, periods of increased correlation with traditional financial markets reduce their diversification benefits.

### Risk Management

The high volatility and drawdowns observed in cryptocurrency markets highlight the importance of appropriate risk management and position sizing.

### Long-Term Investing

Investors seeking long-term capital appreciation may benefit from cryptocurrency exposure, but only if they are willing to tolerate significant short-term fluctuations.

### Asset Allocation

Combining traditional assets such as Gold and the S&P 500 with cryptocurrencies may improve portfolio resilience by balancing growth opportunities and risk.

# Investment Recommendations

Based on the findings of this study:

- Diversify investments across both traditional and digital assets.
- Avoid relying solely on cryptocurrencies for portfolio growth.
- Monitor changes in market correlation during periods of financial uncertainty.
- Consider volatility and drawdown alongside returns when evaluating investment opportunities.
- Rebalance portfolios periodically as market relationships evolve.

# Limitations

- The study relies exclusively on historical price data.
- Correlation analysis measures association and does not establish causality.
- Macroeconomic variables, monetary policy, and geopolitical events were not explicitly modelled.
- Only four representative assets were analysed.
- Cryptocurrency markets continue to evolve, and future relationships may differ.

# Final Conclusion

This project investigated whether cryptocurrencies have decoupled from traditional financial markets by analysing Bitcoin, Ethereum, Gold, and the S&P 500 between 2019 and 2024.

The findings indicate that cryptocurrency market behaviour cannot be described as completely independent from traditional financial markets. Correlation patterns evolved over time, particularly during periods of market stress, suggesting that cryptocurrencies may become more integrated with traditional assets under certain economic conditions.

These results highlight the trade-off between return and risk that investors must consider when incorporating digital assets into diversified portfolios.
